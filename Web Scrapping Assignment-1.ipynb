# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
# Send get request to the webpage server to get the Source code of the page
page=requests.get('https://www.imdb.com/list/ls091520106/')
#Get the page content
soup=BeautifulSoup(page.content)
```

```python
#Pulling all 100 movie names
titles=[]   #Create empty list to store the data
for i in soup.find_all('h3',class_="lister-item-header"):
    titles.append(i.text.replace('\n',''))

#Pulling the ratings of all the movies
ratings=[]   #Create empty list to store the data
for i in soup.find_all('div',class_="ipl-rating-star small"):
    ratings.append(i.text.replace("\n",""))
    
#Pulling the year data of all the movies
year=[]   #Create empty list to store the data
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
#Making Dataframe
df = pd.DataFrame({'Name':titles,'Ratings':ratings[:100],'Year':year})
df
```

```python
#Pulling all 100 movie names
titles=[]   #Create empty list to store the data
for i in soup.find_all('td',class_="titleColumn"):
    titles.append(i.text)

#Pulling the ratings of all the movies
ratings=[]   #Create empty list to store the data
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    ratings.append(i.text.replace('\n',''))

#Pulling the year data of all the movies
year=[]   #Create empty list to store the data
for i in soup.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
    
#Making Dataframe
df = pd.DataFrame({'Name':titles[:100],'Ratings':ratings[:100],'Year':year[:100]})
df
```

# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post

```python
page=requests.get('https://coreyms.com/')
soup=BeautifulSoup(page.content)
```

```python
#Scraping the details of the post Heading, date, content, link for the video

#Initializing the variables
heading=[]
time=[]
content=[]
videoCode = []

#Scrapping the post details
for i in soup.find_all('h2',class_="entry-title"):
    heading.append(i.text)

#Scrap the data of date and time
for i in soup.find_all('time',class_="entry-time"):
    time.append(i.text)

    #Scrap the content 
for i in soup.find_all('div',class_="entry-content"):
    content.append(i.text)

#Scrap the video URL
pageContent = soup.find_all("article")
for i in range (len(pageContent)):

    try:

        videoCode.append(pageContent[i].find_all("iframe")[0].get("src"))

    except:

        videoCode.append("No Video Found")
```

```python
#Creating the dataframe
coreyms_dic={'Heading':heading,'Time':time,'Content':content,"Video_link":videoCode}
coreyms_DataFrame=pd.DataFrame(coreyms_dic)

coreyms_DataFrame
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants')
soup=BeautifulSoup(page.content)
```

```python
#initializing the variables
restaurants=[]    
cuisine=[]
locations=[]
ratings=[]
image=[]
 
    #Scrapping the data
for i in soup.find_all('div',class_="restnt-info cursor"):
    restaurants.append(i.text)
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.strip())
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    locations.append(i.text)
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
for i in soup.find_all('img',class_="no-img"):
    image.append(i["data-src"])

#Creating the data frame
df=pd.DataFrame({'Restaurants':restaurants,'Cuisine':cuisine,'Locations':locations,'Image_URL':image})
df
```

# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
#Initializing the variables
product=[]
price=[]
image=[]

#Scrapping the content of the tags
for i in soup.find_all('div',class_="productCardDetail"):
    product.append(i.text)

for i in soup.find_all('span',class_="discountedPriceText"):
    price.append(i.text)

for i in soup.find_all('img',class_="productImgTag"):
    image.append(i["src"])
```
